[Step1] 데이터 준비 -read_csv() 함수로 자동차 연비 데이터셋 가져오기

```python
# csv 파일을 데이터프레임으로 변환
df = pd.read_csv('auto-mpg.csv', header=None)
```

```python
# 열 이름 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

데이터 살펴보기

```python
df.head()
```

[Step 2] 데이터 탐색 및 데이터 전처리

```python
# 데이터 자료형 확인
df.info()
```

```python
# horsepower 열의 자료형 변경 (문자열 -> 숫자)
df['horsepower'].unique()
```

```python
# '?'을 np.nan으로
df['horsepower'].replace('?',np.nan, inplace=True)
df.info()
```

```python
# 누락 데이터를 삭제
df.dropna(subset=['horsepower'], axis=0, inplace=True)
```

```python
# 문자열 -> 숫자로 변환
df['horsepower'] = df['horsepower'].astype('float')
```

```python
df.info()
```

[Step 3] 속성(feature 또는 variable) 선택

```python
# 분석에 활용할 속성을 선택 (연비, 실린더, 마력, 중량)
ndf = df[['mpg','cylinders','horsepower','weight']]
ndf.head()
```

[Step 4] 데이터셋 구분 - 훈련용(train data)/ 검증용(test data)

```python
# 속성(변수) 선택 
X = ndf[['cylinders','horsepower','weight']] # 독립 변수 X1,X2,X3
y = ndf['mpg'] # 종속 변수 y
```

```python
# train data 와 test data로 구분(7:3 비율)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,                #독립 변수
                                                    y,                #종속 변수
                                                    test_size=0.3,    #검증 30%
                                                    random_state=10)  #랜덤 추출 값
```

[Step 5] Ridge 회귀분석 모형 - sklearn 사용

```python
# sklearn 라이브러리에서 선형회귀분석 모듈 가져오기
from sklearn.linear_model import Ridge

# Lasso  회귀모형 객체 생성 후 학습 시키기
ridge = Ridge().fit(X_train, y_train) # alpha=1.0
```

[Step 6] 성능 평가히기

```python
# 학습을 마친 모형에 test data를 적용하여 결정계수(R-제곱) 계산 -> 결정계수 값이 클수록 모형의 예측 능력이 좋다고 판다
r_square = ridge.score(X_test, y_test)
r_square
```

```python
# 모형에 전체 x 데이터를 입력하여 예측한 값 y_hat으로 담기
y_hat = ridge.predict(X)
```

```python
# 모형에 테스트셋 X_test 데이터를 입력하여 예측한 값 y_hat_test으로 담기
y_hat_test = ridge.predict(X_test)
```

```python
# train 데이터(X_train)의 산점도와 test 데이터로 예측한(y_hat_test) 회귀선을 그래프로 츨력

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
ax.plot(X_train, y_train, 'o', label='Train Data') # 데이터의 분포
ax.plot(X_test, y_hat_test, 'r+', label='Predicted Value') # 모형이 예측한 회귀선
ax.legend(loc='best')
plt.xlabel('features')
plt.ylabel('mpg')
plt.show()
```

```python
print('훈련 세트 정확도 : {:.2f}%'.format(ridge.score(X_train,y_train)*100))
print('테스트 세트 정확도 : {:.2f}%'.format(ridge.score(X_test,y_test)*100))
```

```python
plt.figure(figsize=(10,5))
ax1 = sns.kdeplot(y, label='y')
ax2 = sns.kdeplot(y_hat, label='y_hat', ax=ax1)
plt.title('LinearRegression Ridge 회귀 분석 자동차 연비(MPG) 예측')
plt.legend()
plt.show()
```
